```python
from smolagents import (CodeAgent , tool , LiteLLMModel , OpenAIServerModel)
```

```python
model = OpenAIServerModel(model_id="o4-mini",api_key="")
```

```python
import pandas as pd
```

```python
df = pd.read_csv("jordan_transactions.csv")
```

```python
import pandas as pd
import numpy as np
```

```python
mangaed_agent = CodeAgent(model=model,tools=[],additional_authorized_imports=["math" , "decimal" , "fractions" ,  "numpy" , "scipy" , "sympy" , "numba" , "numpy" , "pandas"],
                          name="data_analysis_agent",
                          description="A data analysis agent that perform data analysis on the provided dataframe",
                          max_steps=20,
                        )
```

```python
manager_agent = CodeAgent(model=model , tools=[],managed_agents=[mangaed_agent],planning_interval=4,max_steps=20,verbosity_level=2)
```

```python
manager_agent.run("You're an expert analyst. You make comprehensive reports and analysis on the provided dataframe that has the following columns: transaction_id: Unique identifier with format JO-DATE-XXXX-XXXXX , .\
mall_name: Name of the mall (Z Mall, C Mall, Y Mall, etc) , branch_name: Specific branch location (Al Bayader, Amman, Gardens, etc.) , transaction_date: Date and time in DD/MM/YYYY HH:MM format ,tax_amount: Tax charged on the transaction \
transaction_amount: Total amount of the transaction , transaction_type: Type of transaction (Sale, Refund) and transaction_status: Status of the transaction (Failed, Completed). answer the following query: calculate summary statistics"
                  ,additional_args={"transaction_dataframe":df},reset=False)
```

```python
from smolagents import GradioUI
```

```python
GradioUI(manager_agent).launch()
```
